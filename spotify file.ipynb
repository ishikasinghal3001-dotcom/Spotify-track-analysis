```python
selected_columns=['popularity','danceability','energy','loudness','tempo','valence','acousticness','instrumentalness']
```

```python
df=df[selected_columns]
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['popularity'],bins=30,kde=True)
plt.title("Distribution of songs popularity")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x="danceability",y="popularity",data=df)
plt.title("Danceability vs Popularity")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df[['danceability','energy','valence']])
plt.title("Audio feature distribution")
plt.show()
```
